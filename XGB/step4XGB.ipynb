```python
import numpy
import pandas
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
```

```python
data_test = pandas.read_csv('../data/features.test.csv').set_index('id')
data_test
```

```python
model = joblib.load('./model/model.joblib')
model
```

```python
predictions = model.predict(
X = data_test.drop(['label'], axis = 'columns')
)
predictions
```

```python
data_test['prediction'] = predictions
data_test
```

```python
confusion_m = sklearn.metrics.confusion_matrix(
y_true = data_test['label'],
y_pred = data_test['prediction'],
)
confusion_m
```

```python
pandas.DataFrame(sklearn.metrics.classification_report(
y_true = data_test['label'],
y_pred = data_test['prediction'],
output_dict = True))
```
