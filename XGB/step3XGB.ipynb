# XGB Step 3 — Train XGBoost model

```python
import pandas as pd
from xgboost import XGBClassifier
import joblib
import json
import sklearn
```

```python
data_train = pd.read_csv('../data/features.train.csv').set_index('id')
data_train
```

```python
model = XGBClassifier(
   n_estimators=2500,
    learning_rate=0.02,
    max_depth=10,
    min_child_weight=3,
    gamma=0.2,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    reg_alpha=0.3,
    random_state=42,
    n_jobs=-1,
    tree_method="hist",
    eval_metric="logloss",
)
```

```python
X_train = data_train.drop(['label'], axis = 'columns')
y_train = data_train['label']
```

```python
model.fit(X_train, y_train)
```

```python
joblib.dump(model, './model/model.joblib')
```

```python
config = model.get_params()
```

```python
with open('./model/config.json', 'w') as f:
    json.dump(config, f, indent=4)
```
