```python
import numpy as np
import pandas as pd
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
```

```python
data_test= pd.read_csv('../data/features.test.csv').set_index('id')
data_test
```

```python
model = joblib.load('./model/model.joblib')
model
```

```python
predictions = model.predict(
X = data_test.drop(['label'], axis = 'columns')
)
predictions
```

```python
data_test['prediction'] = predictions
data_test
```

```python
data_test.to_csv('./results/results.test.csv')
```

```python
confusion_m = sklearn.metrics.confusion_matrix(
y_true = data_test['label'],
y_pred = data_test['prediction'],
)
confusion_m
```

```python
pd.DataFrame(sklearn.metrics.classification_report(
y_true = data_test['label'],
y_pred = data_test['prediction'],
output_dict = True))
```
