```python
import pandas
import joblib
```

```python
data_live= pandas.read_csv('../data/features.live.csv').set_index('id')
data_live
```

```python
model = joblib.load('./model/model.joblib')
model
```

```python
data_live_predictions= pandas.DataFrame(
    data= model.predict(data_live),
   index=data_live.index,
   columns=['predictions']
)
data_live_predictions
```

```python
data_live_predictions.reset_index(inplace=True)
data_live_predictions.to_csv('./results/G7_predictions.live.csv',index=False)
```
