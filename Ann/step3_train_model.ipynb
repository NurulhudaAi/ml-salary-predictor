````python
import joblib
import json
import pandas
import sklearn.neural_network
````

# Instructions

- Read **the train data** from the CSV file and properly set the index


````python
data_train= pandas.read_csv('../data/features.train.csv').set_index('id')
data_train
````

# Instructions

- Instantiate a `sklearn.neural_network.MLPClassifier` model with the following configurations
```
    hidden_layer_sizes    = (20, 10),
    solver                = 'sgd', 
    activation            = 'logistic',
    alpha                 = 0.0,
    batch_size            = 32,
    learning_rate         = 'constant',
    learning_rate_init    = 0.1,
    max_iter              = 1000,
    momentum              = 0.0, 
    nesterovs_momentum    = False, 
    validation_fraction   = 0.0, 
    shuffle               = True,
    early_stopping        = False,
    n_iter_no_change      = 1000,
    random_state          = 0,
```
- Fit the model with data columns properly selected
- Use `joblib.dump` to save the model to `./model/model.joblib`
- Also, use `.get_params`, `json.dump` to save the model configuration to  `./model/config.json`
- Print out all model parameters (not to be confused with model configuration)


````python
model=sklearn.neural_network.MLPClassifier(
    hidden_layer_sizes    = (20, 10),
    solver                = 'sgd', 
    activation            = 'logistic',
    alpha                 = 0.0,
    batch_size            = 32,
    learning_rate         = 'constant',
    learning_rate_init    = 0.1,
    max_iter              = 1000,
    momentum              = 0.0, 
    nesterovs_momentum    = False, 
    validation_fraction   = 0.0, 
    shuffle               = True,
    early_stopping        = False,
    n_iter_no_change      = 1000,
    random_state          = 0,
)
````

````python
X_train = data_train.drop(['label'], axis='columns')
y_train = data_train['label']
````

````python
model.fit(X_train, y_train)
````

````python
joblib.dump(model, './model/model.joblib')
````

````python
config = model.get_params()
````

````python
with open('./model/config.json', 'w') as f:
    json.dump(config, f, indent=4)
````
